```python
import random

def game_engine(count):
	
	is_win = False # Biến lưu trữ người chơi thắng hay thua

	num = random.randint(1, 100)
	
	for i in range (count):
		guess_num = int(input("Nhập số từ 1 đến 100: "))

		if guess_num < 1 or guess_num > 100:
			print("Số bạn chọn nằm ngoài khoảng từ 1 đến 100. Xin vui lòng nhập lại.")
			continue
		
		if num == guess_num:
			print(f"Xin chúc mừng. Bạn đã đoán đúng sau {i + 1} lần. Số được đoán là {num}")
			is_win = True
			break

		else:
			if guess_num < num:
				print(f"Sai rồi, số bạn đoán {guess_num} nhỏ hơn số máy tính đoán")
			else:
				print(f"Sai rồi, số bạn đoán {guess_num} lớn hơn số máy tính đoán")

	print(f"Bạn đã thua cuộc. Sô máy tính chọn là {num}.")
	return is_win

def choose_level():
    print("There are three levels of game:")
    print("\t 1 - Dễ - Bạn có 9 lần để đoán.")
    print("\t 2 - Trung bình - Bạn có 6 lần để đoán.")
    print("\t 3 - Khó - Bạn có 4 lần để đoán.")
    level = 1

    while True:
        level = int(input("Chọn mức độ game: "))
        if 1 <= level <= 3:
            break
    return 9 if level == 1 else 6 if level == 2 else 4

if __name__ == "__main__":
	games_count = 0
	wins_count = 0
	lose_count = 0
	while True:
		games_count += 1 

		game_level = choose_level()
		if game_engine(game_level):
			wins_count += 1
		else:
			lose_count += 1

		response = input("Bạn có muốn chơi lại không? Bấm Enter để chơi tiếp, n/N để thoát: ").lower()
		if response == 'n':
			break

	# In ra thống kê
	print(f"Số lần bạn chơi: {games_count}")
	print(f"Số lần bạn thắng: {wins_count}")
	print(f"Số lần bạn thua: {lose_count}")
	print("Cảm ơn bạn đã chơi!")
```

```python
import random

def game_engine(bet, balance):

    is_win = False

    dice_1 = random.randint(1, 6)
    dice_2 = random.randint(1, 6)
    sum_dice = dice_1 + dice_2
    select = 1

    # Chọn cược
    while True:
        select = int(input("Chọn Tài (1) hay Xỉu (2) hay Cược vào số 5 (3): "))
        if 1 <= select <= 3:
            break

    if select == 1:
        if sum_dice > 5:
            is_win = True
            print("Bạn thắng!")
            balance += bet
        else:
            print("Bạn thua!")
            balance -= bet

    elif select == 2:
        if sum_dice < 5:
            is_win = True
            print("Bạn thắng!")
            balance += bet
        else:
            print("Bạn thua!")
            balance -= bet

    elif select == 3:
        if select == 5:
            is_win = True
            print("Bạn thắng!")
            balance += bet * 3
        else:
            print("Bạn thua!")
            balance -= bet * 3
    
    return is_win, balance



if __name__ == "__main__":
    games_count = 0
    wins_count = 0
    balance = 100000
    bet = 10000

    print(f"Mỗi lần cược là {bet}.")

    while balance >= bet:
        print(f"Bạn đang có {balance}.")
        games_count += 1
        
        is_win, balance = game_engine(bet, balance)
        if is_win:
            wins_count += 1
        
        if balance < bet:
            print("Bạn không còn tiền để chơi tiếp. Trò chơi kết thúc!")
            balance = 0
            break

        response = input("Bạn có muốn tiếp tục chơi? n/N để thoát: ")
        if response == 'n':
            break

print(f"Số ván bạn đã chơi: {games_count}")
print(f"Số ván bạn đã thắng: {wins_count}")
print(f"Số tiền còn lại của bạn: {balance}")
print("Cảm ơn bạn đã chơi!")
```
